Import libraries

```python
import pandas as pd
import numpy as np
import datetime as dt

from pathlib import Path
```

January 2017 file Exploration and Organizing 

```python
file = 'Resources/POSdata/17-01.csv'
df = pd.read_csv(file)
```

```python
df.head()
```

Remove title row from import and rename columns

```python
df = df.drop(labels=0)
df = df.rename(columns={'Unnamed: 1':'Item','Unnamed: 3':'Price','Unnamed: 4':'Sold', 'Unnamed: 7':'Total Sales ($)'})
```

Clean columns

```python
df = df[['Item','Price','Sold','Total Sales ($)']]
df
```

```python
df.info()
```

```python
df.sort_values('Total Sales ($)')
```

Remove rows with NaN values 

```python
df = df.dropna()
df
```

Formatting Total Sales ($) to remove commas and convert from object to float type

```python
df['Total Sales ($)'] = df['Total Sales ($)'].str.replace(r',', '')
df.head()
```

```python
df['Total Sales ($)'] = df['Total Sales ($)'].apply(pd.to_numeric, errors='coerce')
df['Sold'] = df['Sold'].apply(pd.to_numeric, errors='coerce')
df['Price'] = df['Price'].apply(pd.to_numeric, errors='coerce')
df.head()
```

```python
df.sort_values(by =['Total Sales ($)'], ascending=False).tail(20)
```

Adding Month columb to df

```python
df['Month'] = 'Jan'
```

```python
df.head()
```

```python
df.reset_index().tail()
```
